```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
x:list[str] = ['EDA','Data Wrangling','Mineria datos','Limpieza Datos','SQL','Est. Inferencial','ML','DL','IA']
y:list[int] = [5,10,15,20,30,25,31,30,20]
```

```python
data_dict:dict = {
    "caracteristicas":x,
    "Frecuencia" :y
}

data_dict
```

```python
df = pandas.DataFrame(data_dict) #dataframe
```

```python
df
```

```python
from matplotlib import pyplot as plt
df['Frecuencia'].plot(kind='hist', bins=20, title='Frecuencia')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.figure(figsize=(16,9))
sns.barplot(df, x="caracteristicas", y="Frecuencia")
plt.bar(x=x, height=y) # barplot
plt.xticks(rotation=45) # Rotamos etiquetas eje x
plt.ylabel('Frecuencia') # Label y
plt.title('Barplot Caracteristicas') # titulo
plt.tight_layout() # ajuste de ejes
plt.grid(axis='y') # grilla
plt.show()
```
